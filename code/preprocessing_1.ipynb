# 1. PLACE

```python
import pandas as pd
import numpy as np
```

```python
rental_place = pd.read_csv("C:/Users/KimMinyoung/Desktop/2020-1학기/데이터베이스/dataset/서울시공공자전거대여소정보_20181129.csv",encoding='CP949')
```

```python
rental_place
```

```python
place= rental_place.rename(columns={' 구분':'LOCATION_GU', '대여소번호':'PLACE_ID', '대여소명':'PLACE_NAME','거치대수':'BIKE_HOLDER_COUNT','위도':'LATITUDE','경도':'LONGTITUDE'})
place = place.drop(['연번','대여소ID'],axis=1)
place
```

-> place name 앞에 place id가 들어가있다. 따라서 아래와 같이 정규표현식을 사용하여 제거해줄 수 있다.

```python
place['PLACE_NAME']= place['PLACE_NAME'].str.replace(r"[\d.]",'')
```

```python
place
```

# 2. WEATHER

```python
weather = pd.read_excel("C:/Users/KimMinyoung/Desktop/2020-1학기/데이터베이스/dataset/일별기상관측_2018.xlsx",encoding='CP949')
```

```python
weather
```

지점명 columns를 보면 남산을 제외하고는 모두 서울시 00구 이름이다. 따라서 이를 아래와 같이 구를 붙여줄 수 있다.

```python
weather['지점명'] = weather['지점명']+'구'
weather
```

```python
weather.columns
```

```python
weather= weather.rename(columns={'관측일자':'DATE', '지점명':'LOCATION_GU', '평균기온(℃)':'TEMPERATURE','평균습도(%)':'HUMID', '평균풍속(m/s)':'WIND', '강수량(mm)':'PRECIPITATION'})
weather = weather.drop(['최저기온(℃)', '최대기온(℃)', '최저습도(%)','최대풍속(m/s)',
       '최고습도(%)'],axis=1)
```

```python
weather['LOCATION_GU'].unique()
```

```python
weather['LOCATION_GU'][weather['LOCATION_GU']=='중구구'] = '중구'
weather['LOCATION_GU'][weather['LOCATION_GU']=='남산구'] = '남산'
weather['LOCATION_GU'].unique()
```

```python
weather['LOCATION_GU'].value_counts()
```

-> count를 살펴보면 data에 약간의 결측값이 존재하긴 하다는 것을 볼 수 있음

```python
weather
```
